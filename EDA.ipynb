```python
import matplotlib.pyplot as plt
import seaborn as sns
df=sns.load_dataset('titanic')
df.head(10)
```

```python
df.isnull()
```

```python
mean=df['age'].mean()
print(mean)
```

```python
df['Age']=df['age'].fillna(mean)
print(df['Age'])
```

```python
df['deck']=df['deck'].fillna(method='bfill')
print(df['deck'])
```

```python
df['deck'].isnull().sum()
```

```python
df['deck']=df['deck'].fillna(method='pad')
print(df['deck'])
```

```python
df['deck'].isnull().sum()
```
