```python
data={}
data['bbc']=open("/content/bbc (1).txt",'r').read()
data
```

```python
print(data['bbc'])
```

```python
#conversion to lowercase
for k in data.keys():
  data[k]=data[k].lower()
  print(data[k])
```

```python
print(data['bbc'][:800])
```

```python
import re
#Removing panctuation
for k in data.keys():
  data[k]=re.sub(r'[-./?,":;()\']','',data[k])
```

```python
print(data['bbc'][:800])
```

```python
#Removing numbers
for k in data.keys():
  data[k]=re.sub('[-|0-9]','',data[k])#regular expression substitution
```

```python
print(data['bbc'][:800])
```

```python
import nltk
import re
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
#Removing stopwards
import nltk
import re
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords

for k in data.keys():
  data[k]=data[k].split()
  print(data[k])
  stopwords_list=stopwords.words('english')
  stopwords_list=stopwords_list+['mad','max','film','fury','miller','road','the','dust']
  print(stopwords_list)

  for k in data.keys():
    data[k]=[w for w in data[k]if not w in stopwords_list]
    print(data['bbc'])
```

```python
from wordcloud import WordCloud
wordcloud = WordCloud(width = 1000, height =500).generate(' '.join(data['bbc']))
from matplotlib import pyplot as plt
plt.figure(figsize=(15,8))
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```
